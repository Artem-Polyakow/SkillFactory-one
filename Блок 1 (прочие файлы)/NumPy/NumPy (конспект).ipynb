## 2. Биты, байты и хранение чисел

Компьютер хранит данные в виде последовательности `0` и `1`.

Одна позиция, где может быть `0` или `1`, называется **бит**.

8 битов - это **байт**.

Количество возможных комбинаций для `n` бит:

```python
2 ** n
```

Например, в 8 битах:

```python
2 ** 8 = 256
```

То есть 1 байт может хранить 256 разных значений.

Если хранить только неотрицательные целые числа, диапазон будет:

```text
0 ... 255
```

Если хранить отрицательные и положительные числа, диапазон для 8 бит будет:

```text
-128 ... 127
```

Важно понимать: чем больше бит выделено под число, тем больший диапазон значений можно хранить, но тем больше памяти занимает каждое число.

````python
2 ** 8
````


```text
int32:  от -2147483648 до 2147483647

```

Чтобы узнать точные границы целочисленного типа, используется функция:

```python
np.iinfo()
```

````python
np.iinfo(np.int8)
````

````python
print(np.iinfo(np.int16))
print(np.iinfo(np.int32))
print(np.iinfo(np.int64))
````

## 4. Беззнаковые целые типы

Беззнаковые типы хранят только неотрицательные целые числа.

Они называются:

```python
np.uint8
```

## 5. Переполнение типа

Переполнение возникает, когда число не помещается в выбранный тип данных.

Например, `int8` хранит значения только от `-128` до `127`.

Если попытаться поместить туда число больше диапазона, результат может стать неожиданным.

В старых версиях NumPy это часто приводило к "заворачиванию" значения по кругу. Например, для `uint8` значения считаются по модулю 256.

Пример логики:

```python
260 % 256 = 4
```

Поэтому число `260` при приведении к типу с диапазоном 256 значений могло превратиться в `4`.

В современных версиях NumPy часть таких операций может вызывать ошибку переполнения. Но сама идея остаётся важной: **тип данных нужно выбирать так, чтобы все реальные значения помещались в диапазон**.

````python
260 % 256
````

## 7. Почему float может быть неточным

Дробные числа в компьютере хранятся приближённо.

Поэтому не все десятичные дроби можно сохранить идеально точно.

Например, число `4.124` в `float16` может быть округлено до ближайшего доступного значения.

Это особенно важно в задачах:

- финансовых расчётов;
- расчёта комиссий;
- расчёта процентов;
- налогов;
- остатков;
- сверки платежей.

Для учебных и аналитических задач `float32` и `float64` часто подходят. Но для бухгалтерских расчётов лучше использовать специализированные подходы, например хранить деньги в копейках как целые числа или использовать decimal-логику.

````python
print(np.float16(4.12))
print(np.float16(4.13))
print(np.float16(4.123))
print(np.float16(4.124))
print(np.float16(4.125))
````

## 10. Чем массив отличается от списка Python

Массив NumPy:

- хранит элементы одного типа;
- занимает меньше памяти;
- быстрее выполняет массовые числовые операции;
- поддерживает векторизованные вычисления;
- удобен для матриц, векторов и статистики.

Список Python:

- может хранить элементы разных типов;
- гибко меняет длину;
- удобен для общего программирования;
- менее эффективен для больших числовых расчётов.

Важная поправка к теории курса: у Python-списка доступ к элементу по индексу обычно тоже быстрый, потому что Python list реализован как динамический массив. Но NumPy всё равно выигрывает в числовых расчётах, потому что хранит данные компактнее и выполняет операции на уровне оптимизированного кода.

## 12. Тип данных массива

У массива есть атрибут:

```python
dtype
```

Он показывает тип данных элементов массива.

При создании массива можно сразу задать тип:

## 13. Все элементы массива имеют один тип

Это одно из главных правил NumPy.

Если в массив положить значения разных типов, NumPy попытается привести их к одному общему типу.

Например:

```python
[1, 2, 3.5]
```

скорее всего станет массивом с типом `float`.

Если в числовой массив попытаться положить строку, которую можно преобразовать в число, NumPy может её преобразовать.

Если строку нельзя преобразовать в число, возникнет ошибка.

````python
arr = np.array([1, 2, 3.5])

print(arr)
print(arr.dtype)
````

## 14. Основные свойства массива

У массива есть важные атрибуты.

### dtype

Тип данных элементов:

```python
arr.dtype
```

### ndim

Размерность массива:

```python
arr.ndim
```

Одномерный массив имеет `ndim = 1`.

Двумерный массив имеет `ndim = 2`.

### shape

Форма массива:

```python
arr.shape
```

Например, `(3, 4)` означает 3 строки и 4 столбца.

### size

Общее количество элементов:

```python
arr.size
```

Для массива формы `(3, 4)` будет 12 элементов.

### itemsize

Размер одного элемента в байтах:

```python
arr.itemsize
```

Например:

```text
int8  -> 1 байт
int16 -> 2 байта
int32 -> 4 байта
int64 -> 8 байт
```

Общий объём памяти под элементы можно посчитать так:

```python
arr.size * arr.itemsize
```

````python
arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
], dtype=np.int16)

print("dtype:", arr.dtype)
print("ndim:", arr.ndim)
print("shape:", arr.shape)
print("size:", arr.size)
print("itemsize:", arr.itemsize)
print("bytes:", arr.size * arr.itemsize)
````

## 16. np.arange

`np.arange()` создаёт массив чисел с заданным шагом.

Похож на `range`, но работает не только с целыми числами.

Правая граница не включается.

Главное отличие от `range`: `arange` может работать с дробными значениями.

Но с дробными шагами нужно быть осторожным, потому что float-арифметика может давать неточные значения.

````python
print(np.arange(5))
print(np.arange(2.5, 5, 0.5))
````

## 17. np.linspace

`np.linspace()` создаёт заданное количество равномерно распределённых чисел между началом и концом диапазона.

Основные параметры:

```python
np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)
```

Где:

- `start` - начало диапазона;
- `stop` - конец диапазона;
- `num` - сколько чисел нужно создать;
- `endpoint` - включать ли правую границу;
- `retstep` - вернуть ли шаг между значениями;
- `dtype` - тип данных.

Если `endpoint=True`, правая граница включается.

Если `endpoint=False`, правая граница не включается.

`linspace` часто используется:

- для построения графиков;
- для моделирования сценариев;
- для создания равномерных шкал;
- для генерации диапазонов ставок, цен, коэффициентов.

````python
arr, step = np.linspace(1, 2, 10, endpoint=True, retstep=True)

print(arr)
print("step:", step)
````

````python
arr, step = np.linspace(-6, 21, 60, endpoint=True, retstep=True)

print("Шаг с включённой правой границей:", round(step, 2))

arr, step = np.linspace(-6, 21, 60, endpoint=False, retstep=True)

print("Шаг без включения правой границы:", round(step, 2))
````

## 18. Изменение формы массива

Форму массива можно изменить, если общее количество элементов сохраняется.

Например, массив из 8 элементов можно превратить в форму:

```text
(2, 4)
```

или:

```text
(4, 2)
```

Но нельзя превратить в:

```text
(3, 3)
```

потому что там нужно 9 элементов.

## 19. shape и reshape

Можно изменить форму через присваивание `shape`:

```python
arr.shape = (2, 4)
```

Такой способ меняет исходный массив.

Более безопасный и частый способ - использовать:

```python
arr.reshape()
```

Например:

```python
arr.reshape((2, 4))
```

`reshape` возвращает массив новой формы.

Функция, которая позволяет получить новый массив заданной формы, - `reshape`.

````python
arr = np.arange(8)

arr_new = arr.reshape((2, 4))

print("Исходный массив:")
print(arr)

print("Новая форма:")
print(arr_new)
````

## 20. Порядок заполнения при reshape

При изменении формы важно, в каком порядке элементы заполняют новую структуру.

По умолчанию используется порядок:

```python
order='C'
```

Это заполнение по строкам.

Есть также:

```python
order='F'
```

Это заполнение по столбцам.

В большинстве базовых задач используется стандартный порядок `C`.

````python
arr = np.arange(8)

print("order='C':")
print(arr.reshape((2, 4), order='C'))

print("order='F':")
print(arr.reshape((2, 4), order='F'))
````

## 21. Транспонирование

Транспонирование меняет строки и столбцы местами.

Для этого используется:

```python
arr.transpose()
```

или коротко:

```python
arr.T
```

Если была форма `(2, 4)`, после транспонирования станет `(4, 2)`.

Транспонирование часто используется при работе с матрицами, таблицами признаков, расчётами по строкам и столбцам.

````python
arr = np.arange(8).reshape((2, 4))

print(arr)
print(arr.T)
print("shape before:", arr.shape)
print("shape after:", arr.T.shape)
````

## 22. Индексы и срезы

Индексация в NumPy похожа на индексацию списков Python.

Для одномерного массива:

```python
arr[0]
arr[1]
arr[-1]
```

Срезы:

```python
arr[2:5]
arr[:3]
arr[3:]
arr[::-1]
```

Для двумерного массива используется запись через запятую:

```python
matrix[row_index, column_index]
```

Например:

```python
matrix[1, 2]
```

означает элемент во второй строке и третьем столбце.

Можно брать срезы по строкам и столбцам:

```python
matrix[:, 0]      # все строки, первый столбец
matrix[0, :]      # первая строка, все столбцы
matrix[1:3, 2:4]  # строки со второй по третью и столбцы с третьего по четвёртый
```

````python
matrix = np.linspace(0, 6, 12, endpoint=False).reshape(3, 4)

print(matrix)

print("Элемент [1, 2]:", matrix[1, 2])
print("Все строки, первый столбец:", matrix[:, 0])
print("Первая строка:", matrix[0, :])
print("Срез строк и столбцов:")
print(matrix[1:3, 2:4])
````

## 23. Булевы маски

Булева маска - это массив из `True` и `False`, который показывает, какие элементы подходят под условие.

Пример:

```python
arr > 100
```

Вернёт массив:

```text
True, False, True, ...
```

Эту маску можно использовать для фильтрации:

```python
arr[arr > 100]
```

Булевы маски - один из самых важных инструментов NumPy.

Они используются для:

- отбора клиентов;
- поиска аномалий;
- фильтрации продаж;
- отбора строк с нужными условиями;
- замены значений;
- расчёта сегментов.

````python
arr = np.array([50, 120, 300, 80, 500])

mask = arr > 100

print(mask)
print(arr[mask])
````

## 24. Сортировка

Для сортировки есть два способа.

Первый:

```python
np.sort(arr)
```

Он возвращает новый отсортированный массив и не меняет исходный.

Второй:

```python
arr.sort()
```

Он сортирует сам исходный массив и возвращает `None`.

Это важное отличие.

Если нужно сохранить исходный порядок, лучше использовать `np.sort(arr)`.

Если исходный массив больше не нужен в прежнем виде, можно использовать `arr.sort()`.

````python
arr = np.array([23, 12, 45, 12, 23, 4, 15, 3])

arr_sorted = np.sort(arr)

print("Исходный:", arr)
print("Новый отсортированный:", arr_sorted)

result = arr.sort()

print("Результат arr.sort():", result)
print("Исходный после arr.sort():", arr)
````

## 25. Пропущенные значения и np.nan

`np.nan` означает `Not a number`, то есть "не число".

Он появляется, когда результат числовой операции невозможен или значение отсутствует.

Например, квадратный корень из отрицательного числа в действительных числах не существует. В NumPy такая операция может дать `nan`.

Особенность `np.nan`:

```python
np.nan == np.nan
```

даёт:

```text
False
```

Поэтому `nan` нельзя нормально искать через обычное сравнение.

Для проверки используется:

```python
np.isnan()
```

Например:

```python
np.isnan(arr)
```

вернёт булеву маску, где `True` означает, что на этом месте находится `nan`.

````python
data = np.array([4, 9, -4, 3], dtype=np.float64)

roots = np.sqrt(data)

print(roots)
print(np.isnan(roots))
````

## 26. Почему nan опасен в расчётах

Если в массиве есть `nan`, многие итоговые расчёты тоже дадут `nan`.

Поэтому пропуски нужно обрабатывать.

Варианты обработки:

- заменить на 0;
- заменить на среднее;
- заменить на медиану;
- удалить такие значения;
- считать специальными функциями, которые игнорируют `nan`.

В NumPy есть функции:

```python
np.nanmean()
np.nanmedian()
np.nansum()
```

Они считают статистики, игнорируя `nan`.

````python
arr = np.array([1, 2, np.nan, 4])

print("sum:", arr.sum())
print("nansum:", np.nansum(arr))
print("nanmean:", np.nanmean(arr))
print("nanmedian:", np.nanmedian(arr))
````

## 28. Векторные операции

Операции с NumPy-массивами обычно выполняются поэлементно.

Например:

```python
vec1 + vec2
```

означает:

```text
первый элемент + первый элемент
второй элемент + второй элемент
третий элемент + третий элемент
```

То же касается:

```python
vec1 - vec2
vec1 * vec2
vec1 / vec2
vec1 ** 2
```

Векторные операции позволяют избегать ручных циклов и считать быстрее.

````python
vec1 = np.array([2, 4, 7, 2.5])
vec2 = np.array([12, 6, 3.6, 13])

print(vec1 + vec2)
print(vec1 * vec2)
````

```python
arr * 10
```

Если прибавить число:

```python
arr + 100
```

Если возвести в степень:

```python
arr ** 2
```

Это удобно для массовых пересчётов:

- применить коэффициент;
- посчитать НДС;
- рассчитать комиссию;
- нормализовать значения.

````python
arr = np.array([100, 200, 300])

print(arr * 1.05)
print(arr + 100)
print(arr ** 2)
````

## 30. Сравнение массивов

Сравнение массивов тоже происходит поэлементно.

```python
arr > 100
```

вернёт массив из `True` и `False`.

Для нескольких условий используются:

```python
&   # и
|   # или
~   # не
```

Условия нужно брать в скобки:

```python
(arr > 100) & (arr < 500)
```

````python
arr = np.array([50, 120, 300, 800, 40])

mask = (arr > 100) & (arr < 500)

print(mask)
print(arr[mask])
````

## 32. Расстояние между векторами

```python
np.linalg.norm(vec1 - vec2)
```

Это используется, когда нужно понять, насколько похожи два объекта.

## 40. Перемешивание массивов

Для перемешивания исходного массива используется:

```python
np.random.shuffle(arr)
```

Она меняет сам массив и возвращает `None`.

Если нужно получить новый перемешанный массив, не меняя исходный, используется:

```python
np.random.permutation(arr)
```

Это важно в задачах:

- случайное распределение клиентов;
- A/B-тесты;
- перемешивание выборки;
- подготовка данных для обучения модели.

````python
arr = np.arange(10)

permuted = np.random.permutation(arr)

print("Исходный:", arr)
print("Новый перемешанный:", permuted)

result = np.random.shuffle(arr)

print("Результат shuffle:", result)
print("Исходный после shuffle:", arr)
````

## 41. Случайная выборка

Функция:

```python
np.random.choice()
```

позволяет выбрать случайные элементы.

Основные параметры:

```python
np.random.choice(a, size=None, replace=True)
```

Где:

- `a` - массив или число;
- `size` - размер выборки;
- `replace` - можно ли выбирать с повторениями.

Если `replace=True`, один и тот же элемент может попасть в выборку несколько раз.

Если `replace=False`, выборка будет без повторений.

Без повторений обычно выбирают:

- клиентов для обзвона;
- сотрудников для проверки;
- сделки для аудита;
- товары для инвентаризации;
- участников опроса.

С повторениями моделируют процессы, где одинаковый результат может выпадать много раз:

- броски кубика;
- покупки товаров одного типа;
- случайный спрос;
- повторные обращения клиентов;
- количество заказов в день.

````python
clients = np.array([101, 102, 103, 104, 105])

print("Без повторений:", np.random.choice(clients, size=3, replace=False))
print("С повторениями:", np.random.choice(clients, size=10, replace=True))
````

```python
np.random.seed(42)
```

После установки seed случайная последовательность будет повторяться при каждом запуске.

Это важно:

- для сравнения разных решений на одинаковых случайных данных.

````python
np.random.seed(42)

np.random.randint(1, 100, size=10)
````
