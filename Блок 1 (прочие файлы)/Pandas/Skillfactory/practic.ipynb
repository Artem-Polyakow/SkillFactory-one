```python
df_tmp["starttime"] = pd.to_datetime(df_tmp["starttime"])
df_tmp["stoptime"] = pd.to_datetime(df_tmp["stoptime"])

df_tmp["trip duration"] = df_tmp["stoptime"] - df_tmp["starttime"]
df_tmp["trip duration minutes"] = df_tmp["trip duration"].dt.total_seconds() / 60

df_tmp = df_tmp.drop(columns = ["trip duration"])
# df_tmp["trip duration minutes"].value_counts().index[3]
df_tmp.loc[3, "trip duration minutes"]
```

```python
import pandas as pd

df_tmp["starttime"] = pd.to_datetime(df_tmp["starttime"])

def get_time_of_day(hour):
    if 0 <= hour <= 6:
        return "night"
    elif 6 < hour <= 12:
        return "morning"
    elif 12 < hour <= 18:
        return "day"
    else:
        return "evening"

df_tmp["time_of_day"] = df_tmp["starttime"].dt.hour.apply(get_time_of_day)

day_count = df_tmp[df_tmp["time_of_day"] == "day"].shape[0]
night_count = df_tmp[df_tmp["time_of_day"] == "night"].shape[0]

answer = round(day_count / night_count)

print(answer)
```

```python
def make_category(melb_df):
    exclude_cols = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']

    for col in melb_df.columns:
        if col not in exclude_cols and melb_df[col].nunique() <= 150:
            melb_df[col] = melb_df[col].astype('category')

    return melb_df

melb_df = make_category(melb_df)

# melb_df.columns.tolist()

# melb_df.dtypes.value_counts()
```
