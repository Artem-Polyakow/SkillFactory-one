## Короткая шпаргалка

```python
from collections import OrderedDict, deque
```

`OrderedDict` - словарь, который сохраняет порядок добавления ключей. В современных версиях Python обычный `dict` тоже сохраняет порядок, но `OrderedDict` полезен для старых версий и специальных задач.

```python
from collections import OrderedDict

data = [("Ivan", 19), ("Maria", 20), ("Petr", 25)]

ordered_data = OrderedDict(data)

print(ordered_data)
```

`deque` - это двусторонняя очередь. Можно добавлять и удалять элементы с обеих сторон.

```python
from collections import deque

dq = deque()

dq.append("Ivan")      # добавить справа
dq.appendleft("VIP")   # добавить слева

print(dq)

dq.pop()       # удалить справа
dq.popleft()   # удалить слева
