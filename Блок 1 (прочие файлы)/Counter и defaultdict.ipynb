## Короткая шпаргалка перед задачами

````python
from collections import Counter, defaultdict
````

`Counter` нужен, чтобы считать элементы:

````python
colors = ["red", "blue", "red"]

counter = Counter(colors)

print(counter)
print(counter["red"])
print(counter.most_common(1))
````

`defaultdict` нужен, чтобы не проверять каждый раз, есть ли ключ в словаре:


````python
groups = defaultdict(list)

groups[1].append("Ivanov")
groups[1].append("Petrov")

print(groups)
````

## Задача 1. Посчитать, сколько раз встречается каждый товар

Есть список покупок:

```python
products = ["milk", "bread", "milk", "apple", "bread", "milk"]
```

Нужно посчитать, сколько раз купили каждый товар.


### Решение

```python
from collections import Counter

products = ["milk", "bread", "milk", "apple", "bread", "milk"]

product_counter = Counter(products)

print(product_counter)
```

Результат:

```text
Counter({'milk': 3, 'bread': 2, 'apple': 1})
```

### Что происходит

`Counter(products)` проходит по списку и считает, сколько раз встретился каждый элемент.

````python
products = ["milk", "bread", "milk", "apple", "bread", "milk"]

from collections import Counter

product_counter = Counter(products)

print(product_counter)
````

## Задача 3. Найти самый частый товар

Есть список:

```python
products = ["milk", "bread", "milk", "apple", "bread", "milk"]
```

Нужно найти самый популярный товар.

````python
products = ["milk", "bread", "milk", "apple", "bread", "milk"]
product_counter = Counter(products)
print(product_counter.most_common(1))
````

### Решение

```python
from collections import Counter

products = ["milk", "bread", "milk", "apple", "bread", "milk"]

product_counter = Counter(products)

print(product_counter.most_common(1))
```

Результат:

```text
[('milk', 3)]
```

### Как достать только название товара

```python
print(product_counter.most_common(1)[0][0])
```

Результат:

```text
milk
```

### Как достать только количество

```python
print(product_counter.most_common(1)[0][1])
```

Результат:

```text
3
```

Почему так:

```python
product_counter.most_common(1)
```

возвращает список с кортежем:

```python
[('milk', 3)]
```

Значит:

```python
[0]
```

берет первый кортеж:

```python
('milk', 3)
```

А дальше:

```python
[0][0]
```

берет первый элемент кортежа - название товара.


## Задача 4. Посчитать количество уникальных клиентов

Есть список номеров карт:

```python
clients = [101, 102, 101, 103, 102, 101, 104]
```

Нужно узнать, сколько уникальных клиентов было.

````python
clients = [101, 102, 101, 103, 102, 101, 104]
counter_clients = Counter(clients)
print(len(counter_clients))
````

### Решение через Counter

```python
from collections import Counter

clients = [101, 102, 101, 103, 102, 101, 104]

client_counter = Counter(clients)

print(len(client_counter))
```

Результат:

```text
4
```

### Почему так

`Counter` хранит каждый уникальный элемент как ключ:

```python
Counter({101: 3, 102: 2, 103: 1, 104: 1})
```

Значит, `len(client_counter)` показывает количество уникальных ключей.

## Задача 5. Сравнить продажи в двух магазинах

Есть продажи двух магазинов:

```python
shop_1 = ["apple", "banana", "apple", "milk", "milk", "milk"]
shop_2 = ["banana", "banana", "milk", "apple"]
```

Нужно узнать, сколько всего товаров каждого типа продали в двух магазинах вместе.

````python
shop_1 = ["apple", "banana", "apple", "milk", "milk", "milk"]
shop_2 = ["banana", "banana", "milk", "apple"]

counter_shop_1 = Counter(shop_1)
counter_shop_2 = Counter(shop_2)

total_counter = counter_shop_1 + counter_shop_2
print(total_counter)
````

### Решение

```python
from collections import Counter

shop_1 = ["apple", "banana", "apple", "milk", "milk", "milk"]
shop_2 = ["banana", "banana", "milk", "apple"]

counter_1 = Counter(shop_1)
counter_2 = Counter(shop_2)

total_counter = counter_1 + counter_2

print(total_counter)
```

Результат:

```text
Counter({'milk': 4, 'banana': 3, 'apple': 3})
```

### Что происходит

Счетчики можно складывать:

```python
counter_1 + counter_2
```

Python сложит количество по одинаковым ключам.

## Задача 6. Разложить студентов по группам

Есть список студентов:

```python
students = [
    ("Ivanov", 1),
    ("Petrov", 2),
    ("Sidorov", 1),
    ("Smirnov", 3),
    ("Kuznetsova", 2)
]
```

Нужно получить словарь, где ключ - номер группы, а значение - список студентов.

````python
students = [
    ("Ivanov", 1),
    ("Petrov", 2),
    ("Sidorov", 1),
    ("Smirnov", 3),
    ("Kuznetsova", 2)
]

groups = defaultdict(list)

for student, group in students:
    groups[group].append(student)

print(groups)
````

## Задача 7. Собрать уникальные товары по клиентам

Есть покупки:

```python
purchases = [
    ("Ivan", "milk"),
    ("Ivan", "bread"),
    ("Olga", "apple"),
    ("Ivan", "milk"),
    ("Olga", "bread"),
    ("Petr", "milk")
]
```

Нужно получить словарь, где ключ - клиент, а значение - набор уникальных товаров.

````python
purchases = [
    ("Ivan", "milk"),
    ("Ivan", "bread"),
    ("Olga", "apple"),
    ("Ivan", "milk"),
    ("Olga", "bread"),
    ("Petr", "milk")
]

groups = defaultdict(set)

for name, product in purchases:
    groups[name].add(product)

print(groups)
````

## Задача 12. Сравнить продажи двух дней

Есть продажи за два дня:

```python
day_1 = ["coffee", "tea", "coffee", "cake"]
day_2 = ["tea", "tea", "coffee", "juice"]
```

Нужно:

1. Посчитать продажи за каждый день через `Counter`.
    
2. Сложить два счетчика.
    
3. Напечатать общий результат.

````python
day_1 = ["coffee", "tea", "coffee", "cake"]
day_2 = ["tea", "tea", "coffee", "juice"]

count_day_1 = Counter(day_1)
count_day_2 = Counter(day_2)

total_counter = count_day_1 + count_day_2
print(total_counter)
````

## Задача 13. Найти разницу между двумя списками через subtract

```python
month_1 = ["apple", "apple", "banana", "milk", "milk"]
month_2 = ["apple", "banana", "banana", "milk", "bread"]
```

Нужно узнать, каких товаров стало больше или меньше во втором списке относительно первого.

````python
month_1 = ["apple", "apple", "banana", "milk", "milk"]
month_2 = ["apple", "banana", "banana", "milk", "bread"]

count_month_1 = Counter(month_1)
count_month_2 = Counter(month_2)

diff_count = count_month_1.subtract(count_month_2)
print(count_month_1)
````

## Задача 14. Разложить города по странам

cities = [

## Задача 15. Посчитать покупки каждого клиента

Есть список покупок:

```python
purchases = [
    ("Ivan", "milk"),
    ("Olga", "bread"),
    ("Ivan", "apple"),
    ("Ivan", "bread"),
