## Короткая шпаргалка

```python
from collections import OrderedDict, deque
```

`OrderedDict` - словарь, который сохраняет порядок добавления ключей. В современных версиях Python обычный `dict` тоже сохраняет порядок, но `OrderedDict` полезен для старых версий и специальных задач.

```python
from collections import OrderedDict

data = [("Ivan", 19), ("Maria", 20), ("Petr", 25)]

ordered_data = OrderedDict(data)

print(ordered_data)
```

`deque` - это двусторонняя очередь. Можно добавлять и удалять элементы с обеих сторон.

```python
from collections import deque

dq = deque()

dq.append("Ivan")      # добавить справа
dq.appendleft("VIP")   # добавить слева

print(dq)

dq.pop()       # удалить справа
dq.popleft()   # удалить слева
```

# Задачи с подробным разбором

## Задача 1. Создать OrderedDict из списка клиентов

Есть список клиентов:

```python
clients = [
    ("Ivan", 19),
    ("Maria", 22),
    ("Petr", 25)
]
```

Нужно создать `OrderedDict` и вывести его.

````python
from collections import OrderedDict

clients = [
    ("Ivan", 19),
    ("Maria", 22),
    ("Petr", 25)
]

clients_dict = OrderedDict(clients)

print(clients_dict)

````

## Задача 2. Отсортировать клиентов по возрасту

Есть список:

```python
clients = [
    ("Ivan", 19),
    ("Mark", 25),
    ("Andrey", 23),
    ("Maria", 20)
]
```

Нужно создать `OrderedDict`, где клиенты будут идти по возрасту от младшего к старшему.

````python
from collections import OrderedDict

clients = [
    ("Ivan", 19),
    ("Mark", 25),
    ("Andrey", 23),
    ("Maria", 20)
]

clients_dict = OrderedDict(sorted(clients, key = lambda x: x[1]))

print(clients_dict)
````

## Задача 3. Добавить нового клиента в OrderedDict

Есть словарь:

```python
ordered_clients = OrderedDict([
    ("Ivan", 19),
    ("Maria", 20)
])
```

Нужно добавить клиента `"Petr"` с возрастом `25`.

````python
from collections import OrderedDict

ordered_clients = OrderedDict([
    ("Ivan", 19),
    ("Maria", 20)
])

ordered_clients["Petr"] = 25

print(ordered_clients)
````

## Задача 4. Удалить клиента и добавить снова

Есть словарь:

```python
ordered_clients = OrderedDict([
    ("Ivan", 19),
    ("Maria", 20),
    ("Petr", 25)
])
```

Нужно удалить `"Maria"`, а потом добавить ее снова.

````python
from collections import OrderedDict

ordered_clients = OrderedDict([
    ("Ivan", 19),
    ("Maria", 20),
    ("Petr", 25)
])

del ordered_clients["Maria"]

ordered_clients["Maria"] = 21

ordered_clients = OrderedDict(ordered_clients)

print(ordered_clients)

print(sorted(ordered_clients, key = lambda x: x[1]))
````

## Задача 5. Создать очередь клиентов

Есть клиенты:

```python
clients = ["Ivanov", "Petrov", "Sidorov"]
```

Нужно создать очередь и добавить клиентов в конец очереди.

````python
from collections import deque

clients = ["Ivanov2", "Petrov3", "Sidorov"]

clients = deque()

clients.append("Ivanov1")
clients.append("Petrov2")
clients.append("Sidorov")

print(clients)
````

## Задача 6. Обслужить первого клиента из очереди

Есть очередь:

```python
clients = deque(["Ivanov", "Petrov", "Sidorov"])
```

Нужно извлечь первого клиента.

````python
from collections import deque

clients = deque(["Ivanov", "Petrov", "Sidorov"])

first_client = clients.popleft()
two_client = clients.popleft()

print(first_client)
print(two_client)
print(clients)
````

## Задача 7. Добавить VIP-клиента в начало очереди

Есть очередь:

```python
clients = deque(["Ivanov", "Petrov", "Sidorov"])
```

Появился VIP-клиент. Нужно поставить его в начало очереди.

````python
clients = deque(["Ivanov", "Petrov", "Sidorov"])

clients.appendleft("VIP")

print(clients)
````
