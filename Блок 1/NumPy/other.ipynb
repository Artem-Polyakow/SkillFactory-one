```python
import numpy as np


def get_chess(a):
    arr = np.zeros((a, a), dtype=float)

    for i in range(a):
        for j in range(a):
            if (i + j) % 2 == 1:
                arr[i, j] = 1

    return arr

print(get_chess(a = 1))
```

```python
import numpy as np

def shuffle_seed(array):
    seed = np.random.randint(0, 2**32)

    np.random.seed(seed)

    shuffled_array = np.random.permutation(array)

    return shuffled_array, seed

```

```python
import numpy as np

def any_normal(*vectors):
    for i in range(len(vectors)):
        for j in range(i+1, len(vectors)):
            fact = np.dot(vectors[i], vectors[j])
            if fact == 0:
                return True

    return False
```

Напишите функцию get_loto(num), генерирующую трёхмерный массив случайных целых чисел от 1 до 100 (включительно). Это поля для игры в лото.

Трёхмерный массив должен состоять из таблиц чисел формы 5х5, то есть итоговая форма — (num, 5, 5).

Функция возвращает полученный массив.

```python
import numpy as np

import numpy as np

def get_unique_loto(num):
    loto = []

    for i in range(num):
        field = np.random.choice(
            np.arange(1, 101),
            size=(5, 5),
            replace=False
        )
        loto.append(field)

    return np.array(loto)

```
