
## Задача 2. План-факт продаж

```python
import numpy as np

sales_plan = np.array([1_000_000, 1_200_000, 900_000, 1_500_000, 800_000])
sales_fact = np.array([950_000, 1_350_000, 870_000, 1_800_000, 640_000])
```

### Вопросы

1. Посчитай выполнение плана продаж.
2. Найди строки, где план выполнен на 100% и выше.
3. Найди строки, где выполнение ниже 80%.

````python
import numpy as np

sales_plan = np.array([1000000, 1200000, 900000, 1500000, 800000])
sales_fact = np.array([950000, 1350000, 870000, 1800000, 640000])

plan_fact = np.divide(
    sales_fact,
    sales_plan,
    out = np.full(sales_plan.shape, 0, dtype=np.float16),
    where=sales_plan != 0
)*100

print("Выполнение плана продаж -", plan_fact)
print()

good_job = plan_fact >= 100
print("План выполнен на 100 и более % -", plan_fact[good_job])
print()

bad_job = plan_fact < 80
print("План выполнен ниже 80 % -", plan_fact[bad_job])
print()

````

````python
import numpy as np

sales_plan = np.array([1_000_000, 1_200_000, 900_000, 1_500_000, 800_000])
sales_fact = np.array([950_000, 1_350_000, 870_000, 1_800_000, 640_000])

plan_fact = np.divide(
    sales_fact,
    sales_plan,
    out=np.full(sales_plan.shape, 0, dtype=np.float64),
    where=sales_plan != 0
) * 100

print("Выполнение плана продаж (%):", np.round(plan_fact, 2))
print()

good_job = plan_fact >= 100
print("План выполнен на 100%+ (индексы):", np.where(good_job)[0])
print("Значения факт/план:", sales_fact[good_job], "/", sales_plan[good_job])
print("Выполнение (%):", plan_fact[good_job])
print()

bad_job = plan_fact < 80
print("Выполнение ниже 80% (индексы):", np.where(bad_job)[0])
print("Значения факт/план:", sales_fact[bad_job], "/", sales_plan[bad_job])
print("Выполнение (%):", plan_fact[bad_job])
````


## Задача 3. KPI менеджеров

```python
import numpy as np

sales_completion = np.array([0.95, 1.12, 0.97, 1.20, 0.80])
new_clients_completion = np.array([1.00, 0.75, 1.10, 0.90, 0.60])
op_kpi = np.array([0.90, 1.05, 0.85, 1.00, 0.70])
```

Итоговый индекс считается по формуле:

```text
продажи - 50%
новые клиенты - 30%
операционный KPI - 20%
```

### Вопросы

1. Посчитай итоговый индекс для каждого менеджера.
2. Найди менеджеров, у которых индекс ниже 0.85.
3. Найди максимальный и средний индекс.

````python
import numpy as np

sales_completion = np.array([0.95, 1.12, 0.97, 1.20, 0.80])
new_clients_completion = np.array([1.00, 0.75, 1.10, 0.90, 0.60])
op_kpi = np.array([0.90, 1.05, 0.85, 1.00, 0.70])


total_index = sales_completion * 0.5 + new_clients_completion * 0.3 + op_kpi * 0.2
print(total_index)
print()

bad_managers = total_index < 0.85
print(total_index[bad_managers])
print()

max_index = total_index.max()
print("Максимальный индекс ", max_index)
print()

avg_index = total_index.mean()
print("Средний индекс ", avg_index)
print()
````


## Задача 4. Матрица менеджеры x месяцы

```python
import numpy as np

sales_completion = np.array([
    0.90, 1.10, 0.95,
    1.20, 0.80, 1.00,
    0.70, 0.85, 0.95,
    1.05, 1.15, 1.25
])
```

Это выполнение плана 4 менеджеров за 3 месяца.  
Данные идут подряд: сначала 3 месяца первого менеджера, потом 3 месяца второго и так далее.

### Вопросы

1. Преобразуй массив в матрицу формы `(4, 3)`.
2. Посчитай среднее выполнение по каждому менеджеру.
3. Посчитай среднее выполнение по каждому месяцу.

````python
import numpy as np

sales_completion = np.array([
    0.90, 1.10, 0.95,
    1.20, 0.80, 1.00,
    0.70, 0.85, 0.95,
    1.05, 1.15, 1.25
])

matrix = sales_completion.reshape(4, 3)
print("Матрица (менеджеры × месяцы):")
print(matrix)

# axis=1 — среднее по столбцам (месяцам) для каждой строки (менеджера)
avg_by_manager = matrix.mean(axis=1)
print("\nСреднее выполнение по каждому менеджеру:")
for i, val in enumerate(avg_by_manager):
    print(f"  Менеджер {i + 1}: {val:.3f}")

# axis=0 — среднее по строкам (менеджерам) для каждого столбца (месяца)
avg_by_month = matrix.mean(axis=0)
print("\nСреднее выполнение по каждому месяцу:")
for i, val in enumerate(avg_by_month):
    print(f"  Месяц {i + 1}: {val:.3f}")
````


## Задача 5. Маржа и маржинальность

```python
import numpy as np

revenue = np.array([1_200_000, 800_000, 1_500_000, 600_000, 2_000_000])
cost = np.array([900_000, 820_000, 1_100_000, 500_000, 1_850_000])
```

### Вопросы

1. Посчитай маржу по каждой сделке.
2. Посчитай маржинальность по каждой сделке.
3. Найди сделки с отрицательной маржой и сделки с маржинальностью ниже 15%.

````python
import numpy as np

revenue = np.array([1_200_000, 800_000, 1_500_000, 600_000, 2_000_000])
cost = np.array([900_000, 820_000, 1_100_000, 500_000, 1_850_000])

marg = revenue - cost
print(marg)

margin = (revenue - cost) / revenue
print(margin)

bad_deal = (marg < 0) | (margin < 0.15)
bad_indices = np.where(bad_deal)[0]

for i in bad_indices:
    print(f"Сделка {i+1}: маржа={marg[i]:,.0f}₽, маржинальность={margin[i]:.1%}")
````
